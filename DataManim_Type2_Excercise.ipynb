### 서비스 이탈 예측 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
```

```python
# object로 바꿔주기
x_train['HasCrCard'] = x_train['HasCrCard'].astype('object')
x_train['IsActiveMember'] = x_train['HasCrCard'].astype('object')
x_train.info()
```

```python
x_test['HasCrCard'] = x_test['HasCrCard'].astype('object')
x_test['IsActiveMember'] = x_test['HasCrCard'].astype('object')
x_test.info()
```

```python
# Gender 변수 변환
x_train['Gender'] = x_train['Gender'].replace({' male':'Male','female':'Female'})
x_test['Gender'] = x_test['Gender'].replace({' male':'Male','female':'Female'})
```

```python
# 데이터 분리
x = x_train.drop(['CustomerId','Surname'],axis=1)
y = y_train['Exited']

test=x_test.drop(['CustomerId','Surname'], axis=1)
customerId=x_test['CustomerId']
x.head()
```

```python
#인코딩
x=pd.get_dummies(x)
test = pd.get_dummies(test)
x.head()
```

```python
#검증 데이터분리
from sklearn.model_selection import train_test_split
x_train_set, x_val, y_train_set, y_val = train_test_split(x,y,stratify=y, test_size=0.3, random_state=42)
```

```python
#"entropy"
print("roc_auc_score:", roc_auc_score(y_val, y_pred_prob))
```

### 이직여부 판단 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
# 둘 다 결측치가 너무 많네
```

```python
#범주형 데이터 확인
for col in x_train.select_dtypes('object').columns:
    print(f"{col}:{x_train[col].unique()}")
```

```python
#4분위수를 이용한 이상값 확인
for col in x_train.select_dtypes(exclude='object').columns:
    Q1 = x_train[col].quantile(.25)
    Q3 = x_train[col].quantile(.75)
    IQR =Q3 -Q1
    print(col, len(x_train[(x_train[col]<Q1-IQR*1.5) | (x_train[col]>Q3+IQR*1.5)][col]))
```

```python
#마찬가지로 x_test에서도 4분위수를 이용한 이상값 확인
for col in x_test.select_dtypes(exclude='object').columns:
    Q1 = x_test[col].quantile(.25)
    Q3 = x_test[col].quantile(.75)
    IQR =Q3 -Q1
    print(col, len(x_test[(x_test[col]<Q1-IQR*1.5) | (x_test[col]>Q3+IQR*1.5)][col]))
```

```python
# 일단은 캐글에 올라온 내용을 보면서 결측치 처리하기
# 여러개를 보면서 따라해보려고 했으나 결측치 처리하는게 너무 복잡해서 따라 치다가 응용을 못할 것 같아 가장 간단한걸로 먼저 연습하기
# [PY]RF(test AUC:0.7555)
```

```python
# NaN 값이 너무 많이서 일단 최빈값으로 모두 대체
```

```python
x_train.isna().any()
```

```python
#최빈값 구하기
x_train[col].mode()[0]
```

```python
def fill_mode(df):
    for col in df.columns[df.isna().any()].tolist():
        df[col] = df[col].fillna(df[col].mode()[0]) 
    return df
```

```python
x_train = fill_mode(x_train)
x_test = fill_mode(x_test)
print(x_train.isna().sum())
```

```python
#label encoding을 위해 train test 데이터 merge
print(x_train.shape, x_test.shape)
x_df = pd.concat([x_train, x_test], axis=0)
```

```python
object_cols = x_df.select_dtypes(include=object).columns
not_object_cols = x_df.select_dtypes(exclude=object).columns
print(not_object_cols)
```

###

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
def label_encoding(df,cols):
    for col in cols:
        le = LabelEncoder()
        le.fit(df[col])
        x_train[col] = le.transform(x_train[col])
        x_test[col] = le.transform(x_test[col])
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=456)
rf.fit(x_train_sp, y_train_sp)
pred_proba = rf.predict_proba(x_valid_sp)[:, 1]
print(pred_proba)
```
