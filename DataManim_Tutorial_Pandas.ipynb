```python
#Question 57 Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N Less than 40K:a40K - 60K:b60K - 80K:c80K - 120K:d120K +’ : e
```

```python
df.head()
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' :'d',
    '$120K +' : 'e'
}

df['newIncome'] = df['Income_Category'].map(lambda x : dic[x])
```

```python
df['newIncome']
```

```python
#Question 58 Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
```

```python
def changecategory(x):
    if x == 'Unknown':
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    elif x == '$120K +':
        return 'e'
    
df['newIncome'] = df['Income_Category'].apply(changecategory)
```

```python
df['newIncome']
```

```python
#Question 59 Customer_Age의 값을 이ㅛㅇ하여 나이 구간을 AgeState 칼럼으로 정의하라 (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라
```

```python
def agecate(x):
    if 0 <= x < 10:
        return 0
    elif 10 <= x < 20:
        return 10
    elif 20 <= x < 30:
        return 20
    elif 30 <= x < 40:
        return 30
    elif 40 <= x < 50:
        return 40
    elif 50 <= x < 60:
        return 50
    elif 60 <= x < 70:
        return 60
    elif 70 <= x < 80:
        return 70
    
df['AgeState'] = df['Customer_Age'].apply(agecate)
```

```python
df['AgeState'].value_counts().sort_index()
```

```python
# 마님 풀이
df['AgeState'] = df['Customer_Age'].map(lambda x : x//10*10)

ans = df['AgeState'].value_counts().sort_index()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour == '24':
        hour = '00:00:00'

        FinalDate = pd.to_datetime(date + " "+hour) + datetime.timedelta(days=1)
        
    else :
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date + " "+hour)
        
    return FinalDate              
```

```python
#Question 91 df1과 df2를 하나의 데이터 프레임으로 합쳐라
ans = pd.concat([df1, df2])
ans.head()
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
#Question 92 df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘 다 포함하고 있는 녀도에 대해서만 고려한다
```

```python
ans = pd.concat([df3, df4], join='inner')
ans
```

```python
#Question 93 df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 칼럼을 포함하고, 결측치는 0으로 대체한다
```

```python
ans = pd.concat([df3, df4], join='outer').fillna(0)
ans
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
#Question 94 df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하라
```
