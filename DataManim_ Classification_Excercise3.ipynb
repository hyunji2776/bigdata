### 비행탑승 경험 만족도 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv", index_col='ID')
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv", index_col='ID')
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv", index_col='ID')


display(x_train.head())
display(y_train.head())
```

```python
#x_train=x_train.set_index('ID')
#x_train.head()
```

```python
x_train.info()
```

```python
y_train.info()
```

```python
x_test.info()
```

```python
x_test.head()
```

```python
x_train.head()
```

```python
# 둘 다 사용하지 않으므로 'id' 칼럼 지우기
x_train.drop(['id'], axis=1, inplace=True)
x_test.drop(['id'], axis=1,inplace=True)
```

```python
x_train.head()
```

```python
x_train.info()
```

```python
x_train_encoding = x_train[['Gender','Customer Type','Type of Travel','Class']]
```

```python
x_test_encoding = x_test[['Gender','Customer Type','Type of Travel','Class']]
```

```python
temp1 = pd.concat([x_train_encoding, x_test_encoding], axis=0)
temp1_encoding = pd.get_dummies(temp1)
x_train_encoding = temp1_encoding.iloc[:83123]
x_test_encoding = temp1_encoding.iloc[83123:]
```

```python
x_test.info()
```

```python
# x_train과  x_test 둘 다 'Arrival Delay in Minutes' 칼럼에 결측치가 있으므로 중앙값으로 대체해주기
x_train['Arrival Delay in Minutes'] = x_train['Arrival Delay in Minutes'].fillna(x_train['Arrival Delay in Minutes'].median())
```

```python
x_train.info()
```

```python
# x_train과  x_test 둘 다 'Arrival Delay in Minutes' 칼럼에 결측치가 있으므로 중앙값으로 대체해주기
x_test['Arrival Delay in Minutes'] = x_test['Arrival Delay in Minutes'].fillna(x_test['Arrival Delay in Minutes'].median())
```

```python
x_test.info()
```

```python
x_train_encoding.shape, x_test_encoding.shape
```

```python
#원핫인코딩 한 칼럼 삭제해주기
x_train = x_train.drop(['Gender','Customer Type','Type of Travel','Class'], axis=1, inplace=True)
```

```python
x_test = x_test.drop(['Gender','Customer Type','Type of Travel','Class'], axis=1, inplace=True)
```

```python
# 원본 데이터 + 원핫이코딩한 데이터 합쳐주기
x_train = pd.concat([x_train, x_train_encoding], axis=1)
x_test = pd.concat([x_test, x_test_encoding], axis=1)
```

```python
x_test.info()
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
y_train.info()
```

```python
y_train['satisfaction'].value_counts()
```

```python
#y_train이 object형이므로 Label Encoding 해주기
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoder.fit(y_train)
```

### label encoding

```python
y_train['satisfaction']= encoder.fit_transform(y_train['satisfaction'])
```

```python
y_train
```

```python
model = RandomForestClassifier()
result = cross_val_score(model, x_train, y_train)
result.mean()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
```
