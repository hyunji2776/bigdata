```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() # Min-Max-Scale 할 수 있는 객체를 만들어서 반환
scaler.fit(df) #객체를 scaler라는 변수에 저장
df = pd.DataFrame(scaler.transform(df), columns=df.columns)
```

```python
df.head()
```

```python
ans=df[df['qsec'] > 0.5]
```
