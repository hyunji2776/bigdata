```python
## Question56 Temperature의 값이 10미만의 경우 A, 10이상 20미만의 경우 B,20이상 30미만의 경우 C, 그 외의 경우 D라고 할때 각 단계의 데이터 숫자를 구하여라
```

```python
#이렇게는 안됨!
def Temp_level(x):
    if x < 10 :
        x['level']=='A'
    elif 10 < x < 20 :
        x['level']=='B'
    elif 20 < x < 30 :
        x['level']=='C'
    else:
        x['level']=='D'
```

```python
def Temp_level(x):
    if x < 10 :
        return 'A'
    elif 10 < x < 20 :
        return'B'
    elif 20 < x < 30 :
        return'C'
    else:
        return'D'
```

```python
df['Temperature']
```

```python
## Question58 Temperature의 평균이상의 Temperature의 값을 가지는 데이터를 Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?
```

```python
## Question59 Temperature의 중간값 이상의 Temperature의 값을 가지는 데이터를Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?
```

```python
ans = df[df['Temperature'] >= df['Temperature'].median()]
ans=ans.sort_values(by='Temperature')
ans['Humidity'].values[3]
```
