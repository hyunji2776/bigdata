```python
b
```

```python
#IQR 구하기
IQR = b-a
IQR
```

```python
c=IQR*1.5
```

```python
c
```

```python
#1사분위 수  -  IQR*1.5
a=a-c
a
```

```python
#3사분위 수 + IQR*1.5
b=b+c
```

```python
b
```

```python
#이상치들을 구한다
answer = data[(data['bmi'] < a) | (data['bmi'] > b)]
answer.reset_index(drop=True, inplace=True)
answer.head()
```

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
train.head()
```

```python
#Question 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?
```

```python
data = train[train['smoker']=='yes']
data.reset_index(drop=True, inplace=True)
data.head()
```

```python
data=data.sort_values(by='charges', ascending=False)
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
yes=data.iloc[:10]
```

```python
yes2=yes['charges'].mean()
```

```python
no=train[train['smoker']=='no'].sort_values(by='charges',ascending=False).iloc[:10]['charges'].mean()
```

```python
no
```

```python
yes2-no
```

```python
high = train.loc[train.smoker =='yes'].charges.quantile(q=0.90)
high2 = train.loc[train.smoker =='no'].charges.quantile(q=0.90)
mean_yes = train.loc[(train.smoker =='yes') &(train.charges >=high)].charges.mean()
mean_no = train.loc[(train.smoker =='no') &(train.charges >=high2)].charges.mean()
answer = mean_yes - mean_no
print(answer)
```
