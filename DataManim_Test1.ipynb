### 작업 2유형
- 은행의 전화 마케팅에 대해 고객의 반응 여부

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv', index_col='ID')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv', index_col='ID')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv', index_col='ID')

display(train.head())
display(test.head())
display(submission.head())
```

```python
x_train = train.copy()
```

```python
x_test = test.copy()
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
# 인코딩 할 데이터만 따로 만들어주기
x_train_encoding = x_train[['job','marital','education','default','housing','loan','contact','month','poutcome']]
x_train_encoding
```

```python
x_test_encoding = x_test[['job','marital','education','default','housing','loan','contact','month','poutcome']]
x_test_encoding
```

```python
y_train = x_train[['y']]
y_train
```

```python
temp = pd.concat([x_train_encoding, x_test_encoding], axis=0)
temp_encoding = pd.get_dummies(temp)
temp_encoding
```

```python
x_train_encoding = temp_encoding.iloc[:12870]
x_test_encoding = temp_encoding.iloc[12870:]
x_train_encoding.shape, x_test_encoding.shape
```

```python
x_train.drop(['job','marital','education','default','housing','loan','contact','month','poutcome','y'],axis=1, inplace=True)
```

```python
x_train
```

```python
x_test.drop(['job','marital','education','default','housing','loan','contact','month','poutcome'],axis=1, inplace=True)
```

```python
x_test
```

```python
#원데이터랑 합치기
x_train = pd.concat([x_train,x_train_encoding], axis=1)
x_test = pd.concat([x_test,x_test_encoding], axis=1)
```

```python
x_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
model = RandomForestClassifier()
result = cross_val_score(model, x_train, y_train, cv=5)
result.mean()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
```
