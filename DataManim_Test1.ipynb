```python
##Question10 각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?
```

```python
df.head()`
```

```python
df['marital'].unique()
```

```python
ans = df.groupby(['job','marital']).size().reset_index()
ans               
```

```python
pivot_df = ans.pivot_table(index='job',columns='marital')[0]
pivot_df
```

```python
pivot_df = pivot_df.fillna(0)
pivot_df['ratio'] = pivot_df['divorced'] / pivot_df['married']
```

```python
pivot_df
```

```python
result = pivot_df.sort_values('ratio').ratio.values[-1]
print(result)
```
