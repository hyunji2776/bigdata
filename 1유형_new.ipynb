```python
data = pd.read_csv('mtcars.csv')
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.head()
```

```python
#최소최대 척도
scaler = MinMaxScaler()
scaler.fit(data)
data_df = pd.DataFrame(scaler.transform(data), columns=data.columns)
```
