```python
def IQR_rule(val_list):
    Q1 = np.quantile(val_list, 0.25)
    Q3 = np.quantile(val_list, 0.75)
    IQR = Q3 - Q1

    # IQR rule에 위배되지 않는 데이터만 저장
    not_outlier_data = (Q3 + 1.5 * IQR > val_list) & (Q3 - 1.5 * IQR > val_list)
    return not_outlier_data
```
