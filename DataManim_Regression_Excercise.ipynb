### 학생성적 예측 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv", index_col = 'StudentID')
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv", index_col = 'StudentID')
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv", index_col = 'StudentID')


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
x_train.isnull().sum()
```

```python
x_test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
# 인코딩 할 칼럼만 따로 만들어주기 (x_train)
x_train_encoding = x_train[['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery'
                           ,'higher','internet','romantic']]
x_train_encoding.head()
```

```python
# 인코딩 할 칼럼만 따로 만들어주기 (x_test)
x_test_encoding = x_test[['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery'
                           ,'higher','internet','romantic']]
x_test_encoding.head()
```

```python
temp1 = pd.concat([x_train_encoding,x_test_encoding], axis=0)
temp1.head()
```

```python
x_train_encoding.shape[0]
```

```python
x_test_encoding.shape[0]
```

```python
temp1_encoding = pd.get_dummies(temp1)
temp1_encoding
```

```python
x_train_encoding=temp1_encoding.iloc[:678]
x_test_encoding=temp1_encoding.iloc[678:]
x_train_encoding.shape, x_test_encoding.shape
```

```python
x_test_encoding
```

```python
x_train.drop(['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery'
                           ,'higher','internet','romantic'], axis=1, inplace=True)
```

```python
x_test.drop(['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery'
                           ,'higher','internet','romantic'], axis=1, inplace=True)
```

```python
x_train = pd.concat([x_train,x_train_encoding], axis=1)
x_train
```

```python
x_test = pd.concat([x_test,x_test_encoding], axis=1)
x_test
```

```python
x_train.info()
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model = RandomForestRegressor()
result = cross_val_score(model, x_train, y_train)
result.mean()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
pre
```

```python
dic = {'StudentID': x_test.index, 'score':pre}
submit = pd.DataFrame(dic)
submit
```
