### 현대 차량 가격 분류문제 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_train.csv", index_col='ID')
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/y_train.csv", index_col='ID')
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_test.csv", index_col='ID')


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
# 인코딩 할 칼럼
x_train_encoding = x_train[['model','transmission','fuelType']]
x_train_encoding
```

```python
x_test_encoding = x_test[['model','transmission','fuelType']]
x_test_encoding
```

```python
temp1 = pd.concat([x_train_encoding, x_test_encoding], axis=0)
temp1.head()
```

```python
temp1_encoding = pd.get_dummies(temp1)
temp1_encoding
```

```python
x_train_encoding = temp1_encoding.iloc[:3888]
x_test_encoding = temp1_encoding.iloc[3888:]
```

```python
x_train.drop(['model','transmission','fuelType'], axis=1, inplace=True)
x_test.drop(['model','transmission','fuelType'], axis=1, inplace=True)
```

```python
x_train = pd.concat([x_train, x_train_encoding], axis=1)
x_test = pd.concat([x_test, x_test_encoding], axis=1)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
model = RandomForestRegressor()
result = cross_val_score(model, x_train, y_train,)
result.mean()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
```

```python
pre
```

```python
dic = {'ID':x_test.index, 'price':pre}
submit = pd.DataFrame(dic)
submit
```
