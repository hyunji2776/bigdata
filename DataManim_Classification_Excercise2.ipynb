### 정시 배송 여부 판단

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_train.csv", index_col='ID')
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_train.csv", index_col='ID')
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_test.csv", index_col='ID')


display(x_train.head())
display(y_train.head())
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
x_train_encoding = x_train[['Warehouse_block','Mode_of_Shipment','Customer_care_calls','Product_importance','Gender']]
x_train_encoding
```

```python
x_test_encoding = x_test[['Warehouse_block','Mode_of_Shipment','Customer_care_calls','Product_importance','Gender']]
x_test_encoding
```

```python
#인코딩
temp1 = pd.concat([x_train_encoding, x_test_encoding], axis=0)
temp1_encoding = pd.get_dummies(temp1)
temp1_encoding
```

```python
x_train_encoding = temp1_encoding.iloc[:6598]
x_test_encoding = temp1_encoding.iloc[6598:]
```

```python
x_train.drop(['Warehouse_block','Mode_of_Shipment','Customer_care_calls','Product_importance','Gender'], axis=1, inplace=True)
x_test.drop(['Warehouse_block','Mode_of_Shipment','Customer_care_calls' ,'Product_importance','Gender'], axis=1, inplace=True)
```

```python
x_train = pd.concat([x_train, x_train_encoding], axis=1)
x_test = pd.concat([x_test, x_test_encoding], axis=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
x_train.info()
```

```python
model = RandomForestClassifier()
result = cross_val_score(model, x_train, y_train)
result.mean()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
```

```python
pre
```

```python
dic = {'ID':x_test.index, 'Reached.on.Time_Y.N':pre}
submit = pd.DataFrame(dic)
submit
```
